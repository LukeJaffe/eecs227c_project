```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logsumexp(x, a=5):
    return (1 / a) * np.log(np.sum(np.exp(a*x), axis=0))

def logsumexp2(x, a=0.2):
    return a * np.log(np.sum(np.exp(x/a), axis=0))

def smoothmax(x, a=5):
    y = np.sum(x*np.exp(a*x), axis=0) / np.sum(np.exp(a*x), axis=0)
    return y

def normmax(x, p=2, b=0.2):
    """
    Another interesting possibilitiy, which is convex and smooth.
    Could be worth analyzing in addition to logsumexp, especially
    since the norms interact in a potentially interesting way.
    """
    y = (np.sum(x**p, axis=0) + b**p)**(1 / p) - b
    return y
```

```python
_x = np.arange(-1, 1, 0.01)
x = np.vstack([_x, -_x])
y1 = logsumexp(x)
y4 = logsumexp2(x)
y2 = smoothmax(x)
y3 = normmax(x)
print(x.shape, y1.shape, y2.shape, y1.shape)
plt.plot(_x, _x)
plt.plot(_x, -_x)
#plt.plot(_x, y1)
#plt.plot(_x, y2)
plt.plot(_x, y3)
#plt.plot(_x, y4)
#plt.ylim(-0.5, 1)
plt.show()
```
